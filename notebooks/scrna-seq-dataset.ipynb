```python
import matplotlib.pyplot as plt
import pandas as pd
import scanpy as sc
import numpy as np
```

```python
adata = sc.read_h5ad("/kaggle/input/single-cell-rnaseq-data-related-to-cell-cycle/u2os_ann_procpool.h5ad")
```

```python
print(adata)
```

```python
print(adata.obs.head())
```

```python
print(adata.var.head())
```

```python
sc.pp.calculate_qc_metrics(adata, inplace=True)
adata.obs.head()
```

**Remove NAN**

```python
print(adata.obs["cell_cycle_phase"].value_counts(dropna=False))
```

```python
adata = adata[adata.obs['cell_cycle_phase'] != "nan"].copy()
```

```python
print(adata.obs["cell_cycle_phase"].value_counts())
```

```python
adata.obs.head(5)
```

**QC metric distributions**

```python
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'], jitter=0.4, multi_panel=True)
```

**Cell distribution across cell cycle phases**

```python
adata.obs['cell_cycle_phase'].value_counts().plot(kind="bar")
plt.ylabel("Number of cells")
plt.title("Cell distribution across cell cycle phases")
plt.show()
```

**PCA projection of cells**

```python
sc.pl.pca(adata, color=['cell_cycle_phase'], components=['1,2'])
```

**UMAP embedding of cells**

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=30)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['cell_cycle_phase'])
```

**Expression of selected genes**

```python
sc.pl.violin(adata, ['MKI67', 'PCNA', 'CCNB1'], groupby='cell_cycle_phase')
```

**UMAP visualization colored by gene expression**

```python
sc.pl.umap(adata, color=['MKI67', 'PCNA'])
```

**Heatmap of cell cycle gene expression**

```python
s_genes = [
    "MCM5", "PCNA", "TYMS", "FEN1", "MCM2", "MCM4", "RRM1", 
    "UNG", "GINS2", "MCM6", "CDCA7", "DTL", "PRIM1", "UHRF1",
    "HELLS", "RFC2", "RPA2", "NASP", "RAD51AP1", "GMNN"
]

g2m_genes = [
    "HMGB2", "CDK1", "NUSAP1", "UBE2C", "BIRC5", "TPX2", "TOP2A", 
    "NDC80", "CKS2", "NUF2", "CKS1B", "MKI67", "TMPO", "CENPF",
    "TACC3", "FAM64A", "SMC4", "CCNB2", "CKAP2L", "CKAP2"
]

cell_cycle_genes = s_genes + g2m_genes
cell_cycle_genes = [g for g in cell_cycle_genes if g in adata.var_names]

sc.pl.heatmap(adata, var_names=cell_cycle_genes, groupby='cell_cycle_phase', swap_axes=True)
```

**Trajectory-like cell cycle progression**

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=30)
sc.tl.diffmap(adata)

root_cell = np.where(adata.obs['cell_cycle_phase'] == 'G1')[0][0]
adata.uns['iroot'] = root_cell

sc.tl.dpt(adata)

sc.pl.umap(adata, color=['dpt_pseudotime'])
```
